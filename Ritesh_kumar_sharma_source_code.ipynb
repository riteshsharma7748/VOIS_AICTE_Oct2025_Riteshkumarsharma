## Clean data

### Subtask:
Handle missing values and convert data types if necessary.


**Reasoning**:
Drop the 'license' column as it has a high number of missing values.



```python
df = df.drop('license', axis=1)
```

**Reasoning**:
Impute missing values in numerical columns with their respective medians and in categorical columns with their respective modes, and convert 'price', 'service fee', and 'last review' columns to appropriate data types.



```python
numerical_cols = ['reviews per month', 'review rate number', 'minimum nights', 'number of reviews', 'calculated host listings count', 'availability 365', 'host id', 'lat', 'long']
for col in numerical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

categorical_cols = ['NAME', 'host_identity_verified', 'host name', 'neighbourhood group', 'neighbourhood', 'country', 'country code', 'instant_bookable', 'cancellation_policy', 'room type', 'last review', 'house_rules']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

df['price'] = df['price'].astype(str).str.replace(r'[$,]', '', regex=True)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price'] = df['price'].fillna(df['price'].median())

df['service fee'] = df['service fee'].astype(str).str.replace(r'[$,]', '', regex=True)
df['service fee'] = pd.to_numeric(df['service fee'], errors='coerce')
df['service fee'] = df['service fee'].fillna(df['service fee'].median())

df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
df['last review'] = df['last review'].fillna(df['last review'].mode()[0])

df.info()
display(df.isnull().sum())
```

## Analyze data

### Subtask:
Perform some basic statistical analysis and aggregations to understand the distributions and relationships within the data.


**Reasoning**:
Calculate and display descriptive statistics for numerical columns, value counts for categorical columns, and average price and service fee grouped by neighbourhood group and room type to understand the data distribution and relationships.



```python
display(df.describe())
display(df['neighbourhood group'].value_counts())
display(df['room type'].value_counts())
display(df['cancellation_policy'].value_counts())
display(df.groupby(['neighbourhood group', 'room type'])[['price', 'service fee']].mean())
```

## Visualize data

### Subtask:
Create some visualizations to explore the data and present the findings.


**Reasoning**:
Import the necessary libraries for visualization.



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reasoning**:
Create bar plots to visualize the distribution of listings across different 'neighbourhood group' and 'room type' categories.



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='neighbourhood group', palette='viridis')
plt.title('Distribution of Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='room type', palette='viridis')
plt.title('Distribution of Listings by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.show()
```

**Reasoning**:
Create a box plot to visualize the distribution of 'price' for each 'neighbourhood group'.



```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='neighbourhood group', y='price', palette='viridis')
plt.title('Distribution of Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.show()
```

**Reasoning**:
Create a scatter plot to visualize the relationship between 'price' and 'service fee', and add appropriate titles and labels to the plot.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price', y='service fee')
plt.title('Relationship between Price and Service Fee')
plt.xlabel('Price ($)')
plt.ylabel('Service Fee ($)')
plt.show()
```
